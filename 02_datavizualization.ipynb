```python
# Chart 1: Histogram of retweet counts
plt.figure(figsize=(8, 6))
sns.histplot(data=df, x='retweetCount', bins=20)
plt.title('Histogram of Retweet Counts')
plt.xlabel('Retweet Count')
plt.ylabel('Frequency')
plt.show()
```

```python

# Filter the data to include only rows with more than 0 retweets
filtered_data = df[df['retweetCount'] > 0]

# Chart 2: Histogram of retweet counts - Logarithmic scale
plt.figure(figsize=(8, 6))
sns.histplot(data=filtered_data, x='retweetCount', bins=20, log_scale=True)
plt.title('Histogram of Retweet Counts')
plt.xlabel('Retweet Count')
plt.ylabel('Frequency (Log Scale)')
plt.show()
```

```python
# Chart 3: Bar chart of phase counts
plt.figure(figsize=(8, 6))
sns.countplot(data=df, x='phase')
plt.title('Phase Counts')
plt.xlabel('Phase')
plt.ylabel('Count')
plt.show()
```

```python
# Chart 4: Bar chart of Average Sentiment
plt.figure(figsize=(8, 6))
sns.barplot(data=df, x='phase', y='sentiment', estimator=np.mean)
#plt.gca().invert_yaxis()  # Invert the y-axis
plt.title('Average Sentiment by Phase')
plt.xlabel('Phase')
plt.ylabel('Average Sentiment')
plt.show()
```

```python
# Chart 5: Average Retweet, Reply, and Like Counts by Phase
# Group the data by phase and calculate the average counts
avg_counts = df.groupby('phase')[['retweetCount', 'replyCount', 'likeCount']].mean().reset_index()
bar_width = 0.2

# Set the positions of the bars on the x-axis
r1 = np.arange(len(avg_counts))
r2 = [x + bar_width for x in r1]
r3 = [x + bar_width for x in r2]

# Generate the grouped bar chart
plt.figure(figsize=(10, 6))
plt.bar(r1, avg_counts['retweetCount'], color='tab:blue', width=bar_width, label='Retweets')
plt.bar(r2, avg_counts['replyCount'], color='tab:orange', width=bar_width, label='Replies')
plt.bar(r3, avg_counts['likeCount'], color='tab:green', width=bar_width, label='Likes')

# Customize the chart
plt.title('Average Retweet, Reply, and Like Counts by Phase')
plt.xlabel('Phase')
plt.ylabel('Average Count')
plt.xticks([r + bar_width for r in range(len(avg_counts))], avg_counts['phase'])
plt.legend()

# Show the chart
plt.show()
```

```python
# Chart 6: Scatter plot of retweet count vs. reply count
plt.figure(figsize=(8, 6))
sns.scatterplot(data=df, x='retweetCount', y='replyCount')
plt.title('Retweet Count vs. Reply Count')
plt.xlabel('Retweet Count')
plt.ylabel('Reply Count')
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Assuming your dataset is stored in a pandas DataFrame called 'df'

# Calculate average string length by phase
avg_string_length = df.groupby('phase')['content'].apply(lambda x: x.str.len().mean()).reset_index()
avg_string_length.columns = ['phase', 'avg_string_length']

# Bar chart of average string length by phase
plt.figure(figsize=(8, 6))
sns.barplot(data=avg_string_length, x='phase', y='avg_string_length')
plt.title('Average String Length by Phase')
plt.xlabel('Phase')
plt.ylabel('Average String Length')
plt.show()
```
